```python
# Q2. Checking if any numeric value in list using loop

# Solution

# Suppose list 
l1 = ["fruits", "vegetable", "system", 5, "animals", 3, "birds"]

found = False

for item in l1:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("Numeric value exists in the list.")
else:
    print("No numeric value found.")
```

```python
# Q4. Adding all numeric items in dictionary

# Solution

# Example dictionary
d = {"a": "world", "b": 100, "c": "hello", "d": 560}

sum = 0

for value in d.values():
    if isinstance(value, (int, float)):
        sum += value

print("Sum of numeric values:", sum)
```

```python
# Q5. Identifying duplicate values from list

# Solution

# Example List
l1 = [25, 20, 30, 25, 45, 25, 51]

duplicates = []

for item in l1:
    if l1.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```
